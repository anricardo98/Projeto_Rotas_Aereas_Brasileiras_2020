# 3 - Analise da Excentricidade

```python
fig, ax = plt.subplots(1,1,figsize=(10,8))

# layout position
pos = nx.spring_layout(G_2019,seed=123456789,k=0.3)
# color of nodes based on eccentricity
color = list(dict(nx.eccentricity(G_2019)).values())

# draw edges
nx.draw_networkx_edges(G_2019,
                       pos=pos,
                       alpha=0.4, ax=ax)

# draw nodes
nodes = nx.draw_networkx_nodes(G_2019,
                 pos=pos,
                 node_color=color,
                 cmap=plt.cm.jet,ax=ax)

# draw labels
nx.draw_networkx_labels(G_2019, pos=pos,
                        font_color='white', ax=ax)


plt.axis("off")
plt.title('Excentricidade no ano de 2019')
plt.colorbar(nodes)
plt.savefig('eccentricity.png', transparent=True,dpi=300)
plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(10,8))

# layout position
pos = nx.spring_layout(G_2020,seed=123456789,k=0.3)
# color of nodes based on eccentricity
color = list(dict(nx.eccentricity(G_2020)).values())

# draw edges
nx.draw_networkx_edges(G_2020,
                       pos=pos,
                       alpha=0.4, ax=ax)

# draw nodes
nodes = nx.draw_networkx_nodes(G_2020,
                 pos=pos,
                 node_color=color,
                 cmap=plt.cm.jet,ax=ax)

# draw labels
nx.draw_networkx_labels(G_2020, pos=pos,
                        font_color='white', ax=ax)


plt.axis("off")
plt.colorbar(nodes)
plt.title('Excentricidade no ano de 2020')
plt.savefig('eccentricity.png', transparent=True,dpi=300)
plt.show()
```

#4 - Análise de Centralidade 

## 4.1 - Centralidade por grau


```python
# the degree centrality of network(g)
fig, ax = plt.subplots(1,1,figsize=(10,8))

# layout position
pos = nx.spring_layout(G_2019,seed=123456789,k=0.3)
# color of nodes
color = list(dict(nx.degree_centrality(G_2019)).values())

# draw edges
nx.draw_networkx_edges(G_2019,
                       pos=pos,
                       alpha=0.4, ax=ax)

# draw nodes
nodes = nx.draw_networkx_nodes(G_2019,
                 pos=pos,
                 node_color=color,
                 cmap=plt.cm.jet,
                 ax=ax)

# draw labels
nx.draw_networkx_labels(G_2019, pos=pos,
                        font_color='white', ax=ax)


plt.axis("off")
plt.colorbar(nodes)
plt.title('Degree Centrality 2019')
plt.savefig('degree_centrality_2019.png', transparent=True,dpi=300)
plt.show()
```
